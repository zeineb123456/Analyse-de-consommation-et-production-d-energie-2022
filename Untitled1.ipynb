```python
# Convertir la colonne 'Date' en datetime et extraire le nom du mois
def nettoyer_donnees(region_df):
    region_df['Date'] = pd.to_datetime(region_df['Date'], errors='coerce')
    region_df['Mois_nom'] = region_df['Date'].dt.strftime('%B')
    # Calculer la consommation totale par mois (somme de 'Consommation')
    region_df_mois = region_df.groupby('Mois_nom').agg({
        'Consommation': 'sum',
        'Thermique': 'sum',
        'Nucléaire': 'sum',
        'Eolien': 'sum',
        'Solaire': 'sum',
        'Hydraulique': 'sum',
        'Bioénergies': 'sum'
    }).reset_index()
    
    return region_df_mois

# Appliquer la fonction de nettoyage aux DataFrames
bretagne_clean = nettoyer_donnees(bretagne)
iledefrance_clean = nettoyer_donnees(iledefrance)
normandie_clean = nettoyer_donnees(normandie)

# Vérifier le résultat pour une région
print(bretagne_clean.head())
```

```python
# Fusionner les DataFrames des différentes régions
df_combine = pd.merge(bretagne_clean, iledefrance_clean, on='Mois_nom', suffixes=('_bretagne', '_iledefrance'))
df_combine = pd.merge(df_combine, normandie_clean, on='Mois_nom', suffixes=('', '_normandie'))

# Afficher le tableau combiné
print(df_combine.head())
```

```python
def trier_par_mois(df):
    mois_ordre = {
        'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 
        'June': 6, 'July': 7, 'August': 8, 'September': 9, 'October': 10, 
        'November': 11, 'December': 12
    }
    df['Mois_num'] = df['Mois_nom'].map(mois_ordre)  # Associer chaque mois à son numéro
    df = df.sort_values('Mois_num')                 # Trier par numéro de mois
    df = df.drop(columns=['Mois_num'])              # Supprimer la colonne temporaire
    df = df.reset_index(drop=True)                  # Réinitialiser les indices
    return df

# Appliquer la fonction aux 3 tables
bretagne_clean = trier_par_mois(bretagne_clean)
iledefrance_clean = trier_par_mois(iledefrance_clean)
normandie_clean = trier_par_mois(normandie_clean)

# Vérifier les résultats
print(bretagne_clean)
print(iledefrance_clean)
print(normandie_clean)
```
